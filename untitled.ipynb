```python
'''Function to get the RUL of a turbofan engine at a given time cycle'''
def _add_RUL(df):

    # Get the total number of cycles for each unit
    grouped_by_unit = df.groupby(by="unit_no")
    max_cycle = grouped_by_unit["time_cycles"].max()

    # Merge the max cycle back into the original frame
    result_frame = df.merge(max_cycle.to_frame(name='max_cycle'), left_on='unit_no', right_index=True)

    # Calculate remaining useful life for each row
    remaining_useful_life = result_frame["max_cycle"] - result_frame["time_cycles"]
    result_frame["RUL"] = remaining_useful_life

    # drop max_cycle as it's no longer needed
    result_frame = result_frame.drop("max_cycle", axis=1)
    return result_frame

def read_data(filename='dataset/train_FD001.txt',filename_X="dataset/test_FD001.txt",filename_y="dataset/RUL_FD001.txt"):
    """read data from file and create train and test data

    Args:
        filename (str, optional): train filename. Defaults to 'dataset/train_FD001.txt'.
        filename_X (str, optional): test filename. Defaults to "dataset/test_FD001.txt".
        filename_y (str, optional): test filename. Defaults to "dataset/RUL_FD001.txt".
    """
    # read train data and copute RUL and add for training
    train = _add_RUL(pd.read_csv(filename,sep='\s+', header=None, names=col_names))
    test = pd.read_csv(filename_X, sep='\s+',header=None, names=col_names)
    y_test = pd.read_csv(filename_y,sep='\s+', header=None, names=['RUL'])
    return train, test, y_test 
```

```python
def evaluate( y_true, y_hat, label='test'):
    """evaluate model"""
    mse = mean_squared_error(y_true, y_hat)
    rmse = np.sqrt(mse)
    variance = r2_score(y_true, y_hat)
    print(f'{label} set RMSE:{rmse}, R2:{variance}')

# get X_train from train

# Only pass settigns and sensors as features
X_train = train[setting_names +sensor_names].copy()

# get last row of each engine for testing
X_test = test.drop('time_cycles', axis=1).groupby('unit_no').last().copy()

# clip RUL by 125
'''Clipping restricts the values of the RUL (Remaining Useful Life) column to a maximum limit, 
    which is 125 in this case. Any value greater than 125 is replaced with 125, while values less than or equal to 125 remain unchanged.
    
    -> Limit Outliers: Clipping can be used to limit the effect of outliers on the model.
'''
y_train = train['RUL'].copy()
y_train = y_train.clip(upper=125)

# create and fit model
reg = LinearRegression()
reg.fit(X_train.values, y_train.values)

# predict and evaluate
y_pred_train = reg.predict(X_train.values)
evaluate(y_train, y_pred_train, 'train')

y_pred_test = reg.predict(X_test.values)
evaluate(y_test, y_pred_test)
```

#### Applying transformations

```python
'''The add_operating_condition function processes a DataFrame by generating an operating condition (op_cond)
feature based on the engine settings provided in the data.'''
def add_operating_condition(df):
    df_op_cond = df.copy()

    df_op_cond['setting_1'] = df_op_cond['setting_1'].round()
    df_op_cond['setting_2'] = df_op_cond['setting_2'].round(decimals=2)

    # converting settings to string and concatinating makes the operating condition into a categorical variable
    df_op_cond['op_cond'] = df_op_cond['setting_1'].astype(str) + '_' + \
    df_op_cond['setting_2'].astype(str)
    return df_op_cond

'''Scaling the sensor data based on the operating condition of the engine
Four different operating conditions are present in the data, and each condition has a different range of sensor values.
'''
def condition_scaler( df_train, df_test, sensor_names):
    scaler = StandardScaler()
    for condition in df_train['op_cond'].unique():
        scaler.fit(df_train.loc[df_train['op_cond'] == condition, sensor_names])
        df_train.loc[df_train['op_cond'] == condition, sensor_names] = scaler.transform(df_train.loc[df_train['op_cond'] == condition, sensor_names])
        df_test.loc[df_test['op_cond'] == condition, sensor_names] = scaler.transform(df_test.loc[df_test['op_cond'] == condition, sensor_names])
    return df_train, df_test


def exponential_smoothing(df, sensors, n_samples, alpha=0.4):
    """Applies exponential smoothing to sensor data."""
    df = df.copy()
    
    # Apply exponential smoothing and reset the index to align with the original DataFrame
    smoothed = (
        df.groupby('unit_no')[sensors]
        .apply(lambda x: x.ewm(alpha=alpha).mean())
        .reset_index(level=0, drop=True)
    )
    
    # Assign the smoothed values back to the DataFrame
    df[sensors] = smoothed

    # Drop the first `n_samples` of each unit_no to reduce filter delay
    df = df.groupby('unit_no').apply(lambda group: group.iloc[n_samples:]).reset_index(drop=True)
    
    return df
```

```python
# removing all the sensors which were constant visible from the plots
remaining_sensors = ['s_2', 's_3', 's_4', 's_7', 's_8', 's_9',
                        's_11', 's_12', 's_13', 's_14', 's_15', 's_17', 's_20', 's_21']
drop_sensors = [element for element in sensor_names if element not in remaining_sensors]
```

```python
# Clipping the RUL
train['RUL'].clip(upper=125, inplace=True)


X_train = add_operating_condition(train.drop(drop_sensors, axis=1))
X_test = add_operating_condition(test.drop(drop_sensors, axis=1))

X_train, X_test = condition_scaler(X_train, X_test, remaining_sensors)

X_train = exponential_smoothing(X_train, remaining_sensors, 0, 0.4)
X_test = exponential_smoothing(X_test, remaining_sensors, 0, 0.4)
```

```python
def gen_train_data( df, sequence_length, columns):
    '''
    This function generates training data sequences from a DataFrame.
    Iterates over the data, generating subarrays of length sequence_length:
    Start indices: range(0, num_elements - (sequence_length - 1))
    Stop indices: range(sequence_length, num_elements + 1)
    Each subarray is a window of sequence_length rows and len(columns) columns.
    yield is used to return sequences one by one, which can be iterated lazily.
    '''
    data = df[columns].values
    num_elements = data.shape[0]

    # -1 and +1 because of Python indexing
    for start, stop in zip(range(0, num_elements-(sequence_length-1)), range(sequence_length, num_elements+1)):
        yield data[start:stop, :]

def gen_data_wrapper( df, sequence_length, columns, unit_nos=np.array([])):
    '''
    This function wraps around gen_train_data to handle data for multiple units
    For each unit_no: Filters the DataFrame for that unit_no and calls gen_train_data to generate sequences.
    Uses np.concatenate to combine sequences from all units into a single array.
    '''
    if unit_nos.size <= 0:
            unit_nos = df['unit_no'].unique()
    
    data_gen = (list(gen_train_data(df[df['unit_no'] == unit_no], sequence_length, columns))for unit_no in unit_nos)
    data_array = np.concatenate(list(data_gen)).astype(np.float32)
    return data_array

def gen_labels( df, sequence_length, label):
    '''
    This function generates labels corresponding to the sequences created by gen_train_data.
    Extracts the label column as a NumPy array: data_matrix = df[label].values.
    Generates labels starting from the sequence_length - 1 index:
    Each sequence's label is the value of the target column (label) at the last row of the sequence.
    Returns the array of labels.
    '''
    data_matrix = df[label].values
    num_elements = data_matrix.shape[0]

    # -1 because I want to predict the rul of that last row in the sequence, not the next row
    return data_matrix[sequence_length-1:num_elements, :]

def gen_label_wrapper( df, sequence_length, label, unit_nos=np.array([])):
    '''
    This function wraps around gen_labels to handle labels for multiple units.
    
    For each unit_no:
    Filters the DataFrame for that unit_no and calls gen_labels to generate labels.
    Uses np.concatenate to combine labels from all units into a single array.
    Returns the concatenated NumPy array.
    '''
    
    if unit_nos.size <= 0:
            unit_nos = df['unit_no'].unique()
            
    label_gen = [gen_labels(df[df['unit_no'] == unit_no], sequence_length, label)
                    for unit_no in unit_nos]
    label_array = np.concatenate(label_gen).astype(np.float32)
    return label_array

def gen_test_data( df, sequence_length, columns, mask_value):
    '''
    This function generates test sequences, handling cases where the available data is less than the sequence_length.
    If the number of rows in the DataFrame is less than sequence_length:
    Creates a padded array with shape (sequence_length, len(columns)) filled with mask_value.
    Copies the available data into the last rows of the array.
    If the number of rows is greater than or equal to sequence_length:
    Simply extracts the last sequence_length rows.
    Yields the last sequence of length sequence_length.
    '''
    if df.shape[0] < sequence_length:
        data_matrix = np.full(shape=(sequence_length, len(
            columns)), fill_value=mask_value)  # pad
        idx = data_matrix.shape[0] - df.shape[0]
        # fill with available data
        data_matrix[idx:, :] = df[columns].values
    else:
        data_matrix = df[columns].values

    # specifically yield the last possible sequence
    stop = num_elements = data_matrix.shape[0]
    start = stop - sequence_length
    for i in list(range(1)):
        yield data_matrix[start:stop, :]
```

```python
# The sequence_length specifies the length of the time window (i.e., the number of consecutive rows in the time-series data) 
# that will be used as a single input sample for the model.
sequence_length = 20  # default setting in the training function too

# train-val split
gss = GroupShuffleSplit(n_splits=1, train_size=0.80, random_state=42)
for train_unit, val_unit in gss.split(X_train['unit_no'].unique(), groups=X_train['unit_no'].unique()):
    # gss returns indexes and index starts at 1
    train_unit = X_train['unit_no'].unique()[train_unit]
    val_unit = X_train['unit_no'].unique()[val_unit]

    train_split_array = gen_data_wrapper(X_train, sequence_length, remaining_sensors, train_unit)
    train_split_label = gen_label_wrapper(X_train, sequence_length, ['RUL'], train_unit)

    val_split_array = gen_data_wrapper(X_train, sequence_length, remaining_sensors, val_unit)
    val_split_label = gen_label_wrapper(X_train, sequence_length, ['RUL'], val_unit)

# create sequences train, test
train_array = gen_data_wrapper(X_train, sequence_length, remaining_sensors)

label_array = gen_label_wrapper(X_train, sequence_length, ['RUL'])

test_gen = (list(gen_test_data(X_test[X_test['unit_no'] == unit_no], sequence_length, remaining_sensors, -99.))for unit_no in X_test['unit_no'].unique())
test_array = np.concatenate(list(test_gen)).astype(np.float32)
```

#### Building the DCNN Model

```python
input_shape=(sequence_length, train_array.shape[2])
block_size, nblocks = 2, 2
kernel_size=3
dropout=0.5
fc1, fc2 =256, 128
conv_activation='relu' #f.keras.layers.LeakyReLU(alpha=0.1)
batch_normalization=1

input_tensor = Input(input_shape)
x = input_tensor
# DCNN
for i, n_cnn in enumerate([block_size] * nblocks):
    for j in range(n_cnn):
        x = Conv1D(32*2**min(i, 2), kernel_size=kernel_size, padding='same', 
                kernel_regularizer=regularizers.l1_l2(),
                kernel_initializer='he_uniform',
                name='Conv%d_Block%d' % (j, i) )(x)
        if batch_normalization:
            x = BatchNormalization(name='BN%d_Block%d' % (j, i))(x)
        x = Activation(conv_activation, 
                        name='A%d_Block%d' % (j, i))(x)
    x = MaxPooling1D(2, name='MP%d_Block%d' % (j, i))(x)
    if dropout > 0:
        x = Dropout(dropout, name='DO%d_Block%d' % (j, i))(x)

x = Flatten()(x)

# FNN
x = Dense(fc1, name='FC1', 
        kernel_regularizer=regularizers.l1_l2())(x)
x = Activation(conv_activation, name='Act_FC1' )(x)
if dropout > 0:
    x = Dropout(dropout, name='DO_FC1')(x)

x = Dense(fc2, name='FC2', 
        kernel_regularizer=regularizers.l1_l2())(x)
x = Activation(conv_activation, name='Act_FC2')(x)
if dropout > 0:
    x = Dropout(dropout, name='DO_FC2')(x)

x = Dense(1, activation='relu', name='predictions')(x) 

model = Model(inputs=input_tensor, outputs=x)
```

```python
model.summary()
```

```python
model.compile(loss='mean_squared_error', 
                optimizer=Adam(), #lr=0.001, decay=1e-6
                metrics=[tf.keras.metrics.MeanAbsoluteError(name="MAE")])

# Define the EarlyStopping callback
early_stopping = EarlyStopping(
    monitor='val_loss',    # Metric to monitor (can also use 'val_MAE' if preferred)
    patience=20,            # Number of epochs to wait for improvement
    restore_best_weights=True,  # Restore the best weights at the end
    verbose=1              # Print early stopping status
)

# model training
history = model.fit(train_split_array, train_split_label,
                    validation_data=(val_split_array, val_split_label),
                    epochs=100,
                    batch_size=32,
                    callbacks=[early_stopping]
                    )
model.save('model/turbo_RUL_model.h5')
pk.dump(history.history, open("model/history.pk", 'wb'))
```

```python
def plot_loss( fit_history):
    """plot training history

    Args:
        fit_history (_type_): _description_
    """
    plt.figure(figsize=(13, 5))
    plt.plot(range(1, len(
        fit_history.history['loss'])+1), fit_history.history['loss'], label='train')
    plt.plot(range(1, len(fit_history.history['val_loss'])+1),
                fit_history.history['val_loss'], label='validate')
    plt.xlabel('Epochs')
    plt.ylabel('Loss')
    plt.legend()
    plt.show()
    

plot_loss(history)
```
